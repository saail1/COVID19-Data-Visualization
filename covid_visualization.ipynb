# COVID-19 Data Visualization – Pakistan

This notebook analyzes and visualizes COVID-19 cases and deaths in Pakistan using Python libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
pakistan = df[df["location"] == "Pakistan"]
pakistan = pakistan[["date", "total_cases", "new_cases", "total_deaths", "new_deaths"]]
pakistan["date"] = pd.to_datetime(pakistan["date"])
```

```python
# Plot total cases
plt.figure(figsize=(12,6))
sns.lineplot(data=pakistan, x="date", y="total_cases", color='blue')
plt.title("Total COVID-19 Cases in Pakistan")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.grid(True)
plt.show()
```

```python
# Plot daily new cases and deaths
plt.figure(figsize=(12,6))
sns.lineplot(data=pakistan, x="date", y="new_cases", label="New Cases", color='orange')
sns.lineplot(data=pakistan, x="date", y="new_deaths", label="New Deaths", color='red')
plt.title("Daily New COVID-19 Cases and Deaths in Pakistan")
plt.xlabel("Date")
plt.ylabel("Count")
plt.grid(True)
plt.legend()
plt.show()
```
